# Importing Necessary Libraries:

```python
#Import Packages
import cv2 #!pip install opencv-python
import matplotlib.pyplot as plt #!pip install --upgrade opencv-python
```

# Loading the Pre-trained Model and Set-up the configuration:

Note: model weights are downloaded from git repo mentioned above and kept these in the same path of jupyter file from where the program is running.  

```python
#Loading the SSD Model
config_file ='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model ='frozen_inference_graph.pb'
model = cv2.dnn_DetectionModel(frozen_model,config_file)

#Setup the Model configuration
model.setInputSize(320,320)
model.setInputScale(1.0/127.5) ##255/2=127.5
model.setInputMean(127.5) ## Mobilenet input as a [-1,1]
model.setInputSwapRB(True)
```

# Display the Default Class Labels of the Model:

Note: As the pre-trained SSD MobilenetV3 model is trained with coco dataset, and the vehicles I have selected to detect are included in the 80 default classes, I have used the default model to detect 'vehicles' for my project. In my other project, I have used custom dataset (PPE dataset using YOLOv8 model) to train my model as the classes are not in the 
pre-trained model, and also to get better accuracy.

```python
#Defining the class labels
classLabels = [] ##empty list of python
file_name = 'Labels.txt'
with open(file_name,'rt') as fpt:
        classLabels = fpt.read().rstrip('\n').split('\n')
        #classLabels.append(fpt.read())      
print(classLabels)
```

# Detect and Classify Image:

First, I have apply the model to classify image to see whether the model is detecting object correctly or not.

```python
#Read an image detect and classify
img = cv2.imread('2068_Screen_Shot_2019-01-10_at_13.40.48.original.png') ##bgr-blue green red image
plt.imshow(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ClassIndex, confidece, bbox = model.detect(img,confThreshold=0.5)
font_scale =3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, 
                color=(0, 255, 0),thickness=3)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# Object Detection From Recorded Video:

-- This model will only detect vehicles from recorded video including bicycle, car, motorbike, bus & truck. I have tried this model for some of the videos and it is detected different vehicles with a little classification errors.

-- Input is given in cap = cv2.VideoCapture('video path'); type of video could be .mp4, avi or others

-- The accuracy of the model can be increased by fine-tuning 'confThreshold' of the model.

-- The output video will be displayed after runing the code and closed after pressing 'q'. The output video will be saved
afterwards.

```python
#Object Detection from recorded videos
#Input Videos
#cap = cv2.VideoCapture('pexels-george-morina-6107493.MP4')
#cap = cv2.VideoCapture('2022_0101_145528_585.MP4')
#cap = cv2.VideoCapture('2022_0101_151027_590.MP4')
#cap = cv2.VideoCapture('pexels-everett-bumstead-5535905.MP4')
#cap = cv2.VideoCapture('pexels-george-morina-6093814.MP4')
#cap = cv2.VideoCapture('vehicle_detect.mp4')
#Current Input Video
cap = cv2.VideoCapture('test_video.mp4')
cap.set(3,640)
cap.set(4,480)
#Check if the video is open correctly

if not cap.isOpened():
    cap=cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open video")
    
font_scale=3
font=cv2.FONT_HERSHEY_PLAIN

fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('output_vehicle1.mp4', fourcc, 20.0, (640,480))

while True:
    ret, frame = cap.read()
    ClassIndex, confidece, bbox = model.detect(frame,confThreshold=0.50)
    print(ClassIndex)
    
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if (ClassInd==2):
                cv2.rectangle(frame,boxes,(255,0,0),2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, 
                            color=(0, 255, 0),thickness=1)
                
            if (ClassInd==3):
                cv2.rectangle(frame,boxes,(255,0,0),2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, 
                            color=(0, 255, 0),thickness=1)
            
            if (ClassInd==4):
                cv2.rectangle(frame,boxes,(255,0,0),2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, 
                            color=(0, 255, 0),thickness=1)
                
            if (ClassInd==6):
                cv2.rectangle(frame,boxes,(255,0,0),2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, 
                            color=(0, 255, 0),thickness=1)
                                    
            if (ClassInd==8):
                cv2.rectangle(frame,boxes,(255,0,0),2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, 
                            color=(0, 255, 0),thickness=1)
    out.write(frame)       
    cv2.imshow("Object detection Project", frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
      
```
